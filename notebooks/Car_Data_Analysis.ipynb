```python
import pandas as pd

# Import the data
df = pd.read_csv("../data/cab_rides.csv").dropna()
# Shows first 5 lines of the data
df.head()
```

```python
# Shows the last 5 lines of the data
# Can be whatever number like df.tail(30)
df.tail()
```

```python
#Countplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data = df, x = "cab_type", hue = "cab_type")
plt.title("Cab Type Counts")
plt.show()
```

```python
# Distance Histogram

sns.histplot(data = df, x = "distance")
plt.title("Distance Histogram")
plt.show()
```

```python
# Scatterplot
sns.scatterplot(data = df, x = "distance", y = "price")
plt.show()
```

```python
# Boxplot
sns.boxplot(data = df[['distance']])
plt.show()
```

```python
# boxplot for price
sns.boxplot(data = df[['price']])
plt.show()
```

```python
# Prepare input and output
x = df[['distance']]
y = df[['price']]
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the model
model = LinearRegression()

# Model Fitting
model.fit(x,y)

# Model Predictions
predictions = model.predict(x)

# Evaluation
model.score(x,y)
```
